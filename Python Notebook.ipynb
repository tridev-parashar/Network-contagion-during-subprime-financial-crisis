# FINANCIAL CONTAGION BEHAVIOUR

```python
#Load Modules
import networkx as nx
import pandas as pd
import numpy as np
```

```python
#Import datasets - All Data and Pre, During and Post the US subprime crisis of 2008
fcb_alldata = pd.read_csv("Financial Contagion Behaviour - All Data.csv")
fcb_predata = pd.read_csv("Financial Contagion Behaviour - Pre Crisis.csv")
fcb_duringdata = pd.read_csv("Financial Contagion Behaviour - During Crisis.csv")
fcb_postdata = pd.read_csv("Financial Contagion Behaviour - Post Crisis.csv")
```

## 1. Network formation of the entire dataset (All Data)

```python
#All Data
fcb_alldata
```

```python
fcb_alldata.corr(method = 'pearson') #Very high correlation for movements in prices between S&P500 and FTSE100
```

```python
#Pulling out the source nodes, target nodes and weights from the all data dataframe

fcb_alldata2 = fcb_alldata.loc[:,["Source","Target","Weights"]]
fcb_alldata2
```

```python
#Drawing out the network structure for all data dataframe

G_alldata = nx.from_pandas_edgelist(fcb_alldata2,source='Source',target='Target', edge_attr='Weights', create_using=nx.DiGraph())
print(G_alldata)
```

```python
#Drawing out the network structure for pre crisis data df

G_predata = nx.from_pandas_edgelist(fcb_predata2,source='Source',target='Target', edge_attr='Weights', create_using=nx.DiGraph())
print(G_predata)
```

## 5. Calculation of network measures for each of the four networks formed

### 5.1 'All Data' Dataset

```python
#Calculating Outdegree of all data

outdegree1 = G_alldata.out_degree (weight = 'Weights')
G_alldata_od = pd.DataFrame.from_dict(outdegree1)
G_alldata_od.columns = ['Nodes', 'Degree']
G_alldata_od2 = G_alldata_od.sort_values('Degree', ascending = False, ignore_index = True)
G_alldata_od2
```

```python
#Calculating degree distribution
dd1 = G_alldata_od2['Degree'].value_counts(normalize=True)
dd1_df1 = pd.DataFrame (dd1, columns = ['Degree'])
dd1_df2 = dd1_df1.reset_index(level=0)
dd1_df2.columns = ['Degree', 'Distribution of Degree']
dd1_df2
```

```python
#Merging outdegree and distribution dataframes based of degree values

G_alldata_od3 = pd.merge(G_alldata_od2, dd1_df2, on="Degree")
G_alldata_od3
```

```python
#Calculating clustering coefficient, betweenness centrality and closeness centrality of all data

#Clustering Coefficient
cl1 = nx.clustering(G_alldata, weight='Weights')
cl1_df1 = pd.DataFrame(cl1.items(), columns=['Nodes', 'Clustering Coefficient'])

#Betweenness Centrality
bc1 = nx.betweenness_centrality(G_alldata, weight = 'Weights')
bc1_df1 = pd.DataFrame(bc1.items(), columns=['Nodes', 'Betweenness Centrality'])

#Closeness Centrality
cc1 = nx.closeness_centrality(G_alldata, distance = 'Weights')
cc1_df1 = pd.DataFrame(cc1.items(), columns=['Nodes', 'Closeness Centrality'])
```

```python
#Merging clustering, betweenness and closeness with G_alldata_od3

G_alldata_od4 = pd.merge(G_alldata_od3, cl1_df1, on="Nodes")
G_alldata_od5 = pd.merge(G_alldata_od4, bc1_df1, on="Nodes")
G_alldata_od6 = pd.merge(G_alldata_od5, cc1_df1, on="Nodes")
G_alldata_od6
```

```python
#Count number of nodes with '0' clustering coefficient

count1 = (G_alldata_od6['Clustering Coefficient'] == 0).sum()
count1
```

```python
#Finding the node with the highest clustering coefficient

G_alldata_od6[G_alldata_od6['Clustering Coefficient']==G_alldata_od6['Clustering Coefficient'].max()]
```

```python
#Finding the node with the highest and least betweenness centrality

G_alldata_od6[G_alldata_od6['Betweenness Centrality']==G_alldata_od6['Betweenness Centrality'].max()]
```

```python
G_alldata_od6[G_alldata_od6['Betweenness Centrality']==G_alldata_od6['Betweenness Centrality'].min()]
```

```python
#Finding the node with the highest and least closeness centrality
G_alldata_od6[G_alldata_od6['Closeness Centrality']==G_alldata_od6['Closeness Centrality'].max()]
```

```python
G_alldata_od6[G_alldata_od6['Closeness Centrality']==G_alldata_od6['Closeness Centrality'].min()]
```
